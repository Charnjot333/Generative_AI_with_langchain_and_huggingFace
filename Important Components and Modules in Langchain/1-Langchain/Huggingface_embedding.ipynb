## Sentence Transformers on Hugging Face 
Sentence Transformers is a Python library developed by Hugging Face.

Instead of generating embeddings for individual words, it generates one embedding for the entire sentence.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "This is the text "
query_results = embedding.embed_query(text)
query_results
```
